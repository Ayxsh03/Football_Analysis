```python
from roboflow import Roboflow
rf = Roboflow(api_key="YOUR API KEY HERE")
project = rf.workspace("roboflow-jvuqo").project("football-players-detection-3zvbc")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
import shutil
import os

src_base = os.path.join(dataset.location)
dst_base = os.path.join(dataset.location, 'football-players-detection-1')

shutil.move(os.path.join(src_base, 'train'), os.path.join(dst_base, 'train'))
shutil.move(os.path.join(src_base, 'test'), os.path.join(dst_base, 'test'))
shutil.move(os.path.join(src_base, 'valid'), os.path.join(dst_base, 'valid'))
```
